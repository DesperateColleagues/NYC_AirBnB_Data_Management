```python
Subway_stops = df.groupby(["stop_name", "stop_lat", "stop_lon"]).first().reset_index()
Subway_stops = Subway_stops.loc[:, 'stop_name':'stop_id']
```

```python
Subway_stops
```

```python
Subway_stops = (
    Subway_stops
        .rename(columns={
            'stop_name': 'name',
            'stop_lat': 'latitude',
            'stop_lon': 'longitude',
            'stop_id': 'id'
        }).reindex(columns=['id', 'name', 'latitude', 'longitude'])
)

Subway_stops
```

```python
Subway_stops.to_csv('../out/subway_stops.csv', index=False)
```
